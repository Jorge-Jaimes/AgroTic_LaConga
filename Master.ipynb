```python
def ListasFrames(List_IDDevice,caracter):
  """This function make a list of al the dataframes values
  """
  ListDispositivos = []
  nameDev = []
  for i in List_IDDevice.keys():
    if List_IDDevice[i]["rango_datos"][0]==0 and List_IDDevice[i]["rango_datos"][1]==0:
      del List_IDDevice[i]
    else:
      if i[:1]==caracter:
        data = MakeSens.download_data(i,List_IDDevice[i]["rango_datos"][0],List_IDDevice[i]["rango_datos"][1],frecuency,token)
        list1 = ["temperatura", "humedad", "presion"]
        result =  all(elem in data.columns  for elem in list1)
        if result: 
          ListDispositivos.append(data) #Tienes los datos en un DataFrame
          nameDev.append(i)
          if len(ListDispositivos)!=1:
            temp = np.intersect1d(temp, data.columns)
          else: 
            temp = data.columns
          
  return ListDispositivos, temp, nameDev
```

```python
def transDF(Dispositivos):
  """This function transform to one dataframe
  """
  for j in range(len(Dispositivos)-1):
    if j==0:
      DF = Dispositivos[j]
    else:
      DF = DF.append(Dispositivos[j+1])
  return DF
```

```python
DfR = transDF(DispositivosR)
DfE = transDF(DispositivosE)
```

```python
def transDdf(data):
  """This function transform the dataframe to days
  """
  data.index = pd.to_datetime(data.index)
  data = data.reset_index(level=0)
  data.rename(columns = {data.columns[0]: 'Fecha'}, inplace=True)
  dataF = pd.DataFrame()
  for i, j in enumerate(data.Name_Device.unique()):
    df1 = data.loc[data['Name_Device'] == j] 
    df1.index = df1.Fecha
    df1 = df1.drop('Fecha', axis=1)
    df1 = df1.resample('D').mean()
    df1["Name"] = j 
    dataF = dataF.append(df1)
  return dataF
```
